```python
# df.append(df): í–‰ ì¶”ê°€í•˜ê¸°
data={'ë´„':np.nan,'ì—¬ë¦„':np.nan,'ê°€ì„':23,'ê²¨ìš¸':9}
df_weather.append(data, ignore_index=True)
```

```python
# df['ì¹¼ëŸ¼ëª…']=seq_data: ì—´ ì¶”ê°€í•˜ê¸°
df_weather['ì—°ë„']=[2017,2018,2019,2020]
df_weather=df_weather[['ì—°ë„','ë´„','ì—¬ë¦„','ê°€ì„','ê²¨ìš¸']]
df_weather
```

#### (3) ë°ì´í„°í”„ë ˆì„ í•©ì¹˜ê¸°

```python
# df.merge(df2, how='ì¡°ì¸ ë°©ë²•', on='ì¹¼ëŸ¼ëª…'): ë‹¤ë¥¸ ë°ì´í„°í”„ë ˆì„ì´ë‘ í•©ì¹˜ê¸°
df_precipitation=pd.DataFrame([{'ì—°ë„':2017,'ê°•ìˆ˜ëŸ‰':1230},{'ì—°ë„':2019,'ê°•ìˆ˜ëŸ‰':1524}])

# ë‚´ë¶€ì¡°ì¸: nanê°’ì´ ìƒê¸°ë©´ ê²°í•©X
df_weather.merge(df_precipitation,how='inner',on='ì—°ë„')
```

#### (4) ë°ì´í„°í”„ë ˆì„ ì „ì²˜ë¦¬í•˜ê¸°

```python
# df.isnull(): ë°ì´í„°ì— nullê°’ì´ ìˆìœ¼ë©´ True ì—†ìœ¼ë©´ Falseë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤
df_weather.isnull().sum()
```

```python
# df.fillna(): nullê°’ì„ ì±„ì›Œì¤ë‹ˆë‹¤
df_weather['ë´„'].fillna(df_weather['ë´„'].mean(), inplace=True) #inplace=TrueëŠ” ì›ë³¸ë°ì´í„°ë¥¼ ë°”ê¿”ì¤ë‹ˆë‹¤
df_weather
```

```python
# df.drop('ì¹¼ëŒëª…',axis=): ë°ì´í„°ë¥¼ ì‚­ì œí•©ë‹ˆë‹¤
df_weather.drop('ê°•ìˆ˜ëŸ‰',axis=1,inplace=True)
df_weather
```

```python
# df.sort_values(by='ì¹¼ëŸ¼ëª…', ascending=True): ì¹¼ëŸ¼ëª… ê¸°ì¤€ìœ¼ë¡œ ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬í•©ë‹ˆë‹¤
df_weather.sort_values(by='ê²¨ìš¸')
```
